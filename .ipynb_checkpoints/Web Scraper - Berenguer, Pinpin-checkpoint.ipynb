```python
import requests
import pandas as pd
import json
```

```python
URL = "https://www.dlsu.edu.ph"
page = requests.get(URL)
```

```python
from bs4 import BeautifulSoup
import time

soup = BeautifulSoup(page.text, 'html.parser')
 
urls = []
start_time = time.time()

for link in soup.find_all('a'):
    get_link = link.get('href')
    if(not (get_link.startswith("https://") or get_link.startswith("http://"))):
        get_link = "https://www.dlsu.edu.ph" + str(link.get('href'))
    
    urls.append(get_link)
        
duration = time.time() - start_time
```

### List of All URLs with duplicates

```python
print('List of All URLs')
print('\n'.join(urls))
```

```python
print('Elapsed time: ' + str(duration))
print('Total URLs collected: ' + str(len(urls)))
```

```python
def remove_duplicates(links):
    for url in links:
        if (url + str('/')) in links:
            print(url)
            links.remove(url)
    links = list(dict.fromkeys(links))
    return links

urls = remove_duplicates(urls)
```

```python
print('List of All URLs')
print('\n'.join(urls))
```

```python
print('List of All URLs without duplicates: ' + str(len(urls)))
```

```python
import re
from selenium import webdriver

#  A function that returns all the emails from any String
def get_emails(url):
    chrome_driver = 'D:\Documents\webdriver\chromedriver'
    driver = webdriver.Chrome(chrome_driver)
    driver.get(url)

    page_source = driver.page_source

    #  Used to recognise emails in text
    email_regex = r"""(?:[a-z0-9!#$%&'*+/=?^_`{|}~-]+(?:\.[a-z0-9!#$%&'*+/=?^_`{|}~-]+)*|"(?:[\x01-\x08\x0b\x0c\x0e-\x1f\x21\x23-\x5b\x5d-\x7f]|\\[\x01-\x09\x0b\x0c\x0e-\x7f])*")@(?:(?:[a-z0-9](?:[a-z0-9-]*[a-z0-9])?\.)+[a-z0-9](?:[a-z0-9-]*[a-z0-9])?|\[(?:(?:(2(5[0-5]|[0-4][0-9])|1[0-9][0-9]|[1-9]?[0-9]))\.){3}(?:(2(5[0-5]|[0-4][0-9])|1[0-9][0-9]|[1-9]?[0-9])|[a-z0-9-]*[a-z0-9]:(?:[\x01-\x08\x0b\x0c\x0e-\x1f\x21-\x5a\x53-\x7f]|\\[\x01-\x09\x0b\x0c\x0e-\x7f])+)\])"""

    
    list_of_emails = []

    for re_match in re.finditer(email_regex, page_source):
        list_of_emails.append(re_match.group())

    return list_of_emails


#  Grabbing the emails from the sample text
emails = []
emails = get_emails('https://www.dlsu.edu.ph/colleges/ccs/office-of-the-dean/')
print('\n'.join(emails))
```
